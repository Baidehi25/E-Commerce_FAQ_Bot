```python
from sentence_transformers import SentenceTransformer
import chromadb
embedder=SentenceTransformer('all-MiniLM-L6-v2')
client=chromadb.Client()
collection=client.get_or_create_collection("ecommerce_faq")
texts = []
ids = []
metadatas = []
for kb in KnowledgeBase:
    ids.append(kb["id"])
    texts.append(kb["text"])
    metadatas.append({"topic": kb["topic"]})
embeddings = embedder.encode(texts).tolist()
collection.add(
    documents=texts,
    embeddings=embeddings,
    ids=ids,
    metadatas=metadatas
)
```

```python
from typing import TypedDict, List
class CapstoneState(TypedDict):
    question: str
    messages: List[dict]
    route: str
    retrieved: str
    sources: List[str]
    tool_result: str
    answer: str
    faithfulness: float
    eval_retries: int
    user_name: str
print("state defined:", list(CapstoneState.__annotations__.keys()))
```

```python
def memory_node(state: CapstoneState) -> dict:
    messages = state.get("messages", [])
    user_name = state.get("user_name", "")

    messages.append({"role": "user", "content": state["question"]})
    messages = messages[-6:]

    if "my name is" in state["question"].lower():
        parts = state["question"].lower().split("my name is")
        if len(parts) > 1:
            user_name = parts[1].strip().split()[0].rstrip('.,!?;:').capitalize()

    return {"messages": messages, "user_name": user_name}

result=memory_node({"question": "My name is Priya. Help me.", "messages": [], "user_name": ""})
print("name found:", result["user_name"])
print("works" if result["user_name"]=="Priya" else "something is wrong")
```

```python
from langchain_groq import ChatGroq

llm=ChatGroq(model="llama-3.3-70b-versatile", temperature=0)

def router_node(state: CapstoneState) -> dict:
    prompt=f"""You are a router for an e-commerce FAQ chatbot.
Decide how to answer the question below.

Routes:
- retrieve: question is about returns, shipping, payments, orders, tracking, coupons, warranty, exchanges, or account
- tool: question needs today's date or a delivery date calculation
- memory_only: greeting like hi or hello, thanks, or casual chitchat

Question: {state["question"]}

Reply with one word only: retrieve, tool, or memory_only"""

    response=llm.invoke(prompt)
    route=response.content.strip().lower().replace(".", "").replace(",", "")

    if route not in ["retrieve", "tool", "memory_only"]:
        route="retrieve"  # default if the llm gives unexpected output

    print(f"route: {route}")
    return {"route": route}

print("test 1 (expect retrieve):")
router_node({"question": "What is your return policy?"})

print("test 2 (expect tool):")
router_node({"question": "When will my order arrive if I order today?"})

print("test 3 (expect memory_only):")
router_node({"question": "Hello!"})
```

```python
def retrieval_node(state: CapstoneState) -> dict:
    query_embedding=embedder.encode([state["question"]]).tolist()
    results=collection.query(query_embeddings=query_embedding, n_results=3)

    chunks=[]
    sources=[]
    for doc, meta in zip(results["documents"][0], results["metadatas"][0]):
        chunks.append(f"[{meta['topic']}]\n{doc.strip()}")
        sources.append(meta["topic"])

    retrieved="\n\n---\n\n".join(chunks)
    print(f"sources found: {sources}")
    return {"retrieved": retrieved, "sources": sources}

# quick test
result=retrieval_node({"question": "How do I cancel my order?"})
print("\nfirst 200 characters retrieved:")
print(result["retrieved"][:200])
```

```python
def skip_retrieval_node(state: CapstoneState) -> dict:
    return {"retrieved": "", "sources": []}
```

```python
from datetime import datetime, timedelta

def tool_node(state: CapstoneState) -> dict:
    try:
        today=datetime.now()
        standard=today + timedelta(days=7)
        express=today + timedelta(days=3)

        if today.hour < 12:
            same_day_info="available today if you order now"
        else:
            next_day=(today + timedelta(days=1)).strftime("%d %B %Y")
            same_day_info=f"available tomorrow: {next_day}"

        tool_result=f"""today's date: {today.strftime("%A, %d %B %Y")}
standard shipping arrives by: {standard.strftime("%A, %d %B %Y")} (5-7 days)
express shipping arrives by: {express.strftime("%A, %d %B %Y")} (2-3 days)
same-day delivery: {same_day_info}"""

        return {"tool_result": tool_result, "retrieved": "", "sources": ["Date Calculator"]}

    except Exception as e:
        return {"tool_result": f"could not get date info: {str(e)}", "retrieved": "", "sources": []}

result=tool_node({"question": "when will my order arrive?"})
print(result["tool_result"])
```

```python
def answer_node(state: CapstoneState) -> dict:
    context_parts=[]
    if state.get("retrieved"):
        context_parts.append(f"KNOWLEDGE BASE:\n{state['retrieved']}")
    if state.get("tool_result"):
        context_parts.append(f"TOOL RESULT:\n{state['tool_result']}")
    context="\n\n".join(context_parts)

    if state.get("user_name"):
        name_line=f"Address the customer as {state['user_name']}."
    else:
        name_line=""

    if state.get("eval_retries", 0) > 0:
        retry_line="Important: your previous answer was flagged. Use only the context below. No outside knowledge."
    else:
        retry_line=""

    system_prompt=f"""You are ShopEasy's customer support assistant. {name_line}
{retry_line}

Rule: answer only using information from the context below.
If the answer is not in the context, say: I do not have that information. Please call 1800-123-4567.
Never make up facts. Never use your general knowledge.

Context:
{context if context else "no context. for greetings and chitchat, respond warmly and ask how you can help."}"""

    messages=[{"role": "system", "content": system_prompt}]
    messages.extend(state.get("messages", []))

    response=llm.invoke(messages)
    answer=response.content.strip()
    print(f"answer generated ({len(answer)} characters)")
    return {"answer": answer}

test_state={
    "question": "Can I return something?",
    "messages": [{"role": "user", "content": "Can I return something?"}],
    "retrieved": "[Return Policy]\n30-day return. Items must be unused in original packaging.",
    "tool_result": "",
    "eval_retries": 0,
    "user_name": ""
}
result=answer_node(test_state)
print(result["answer"][:200])
```

```python
import re

FAITHFULNESS_THRESHOLD=0.7
MAX_EVAL_RETRIES=2

def eval_node(state: CapstoneState) -> dict:
    if not state.get("retrieved") and not state.get("tool_result"):
        return {"faithfulness": 1.0, "eval_retries": state.get("eval_retries", 0)}
    answer_lower=state.get("answer", "").lower()
    honest_responses=["do not have that information", "don't have that information", "not in my knowledge"]
    if any(phrase in answer_lower for phrase in honest_responses):
        return {"faithfulness": 1.0, "eval_retries": state.get("eval_retries", 0)}

    context=state.get("retrieved", "") or state.get("tool_result", "")

    eval_prompt=f"""Rate the faithfulness of this answer from 0.0 to 1.0.
            Faithfulness means: does the answer use only information from the context?
            Context: {context}
            Answer: {state["answer"]}
            1.0 = answer only uses the context
            0.5 = mixes context with outside knowledge
            0.0 = ignores the context completely
            Reply with a number only, for example: 0.85"""
    try:
        response=llm.invoke(eval_prompt)
        numbers=re.findall(r'\d+\.?\d*', response.content.strip())
        score=float(numbers[0]) if numbers else 0.5
        score=max(0.0, min(1.0, score))
    except Exception:
        score=0.5
    retries=state.get("eval_retries", 0) + 1
    print(f"faithfulness: {score:.2f}, retries so far: {retries}")
    return {"faithfulness": score, "eval_retries": retries}
```

```python
def save_node(state: CapstoneState) -> dict:
    messages=state.get("messages", [])
    messages.append({"role": "assistant", "content": state["answer"]})
    print("answer saved to history")
    return {"messages": messages}
```

```python
from langgraph.graph import StateGraph, END
from langgraph.checkpoint.memory import MemorySaver

def route_decision(state: CapstoneState) -> str:
    route=state.get("route", "retrieve")
    if route=="tool":
        return "tool"
    elif route=="memory_only":
        return "skip"
    else:
        return "retrieve"

def eval_decision(state: CapstoneState) -> str:
    faith=state.get("faithfulness", 0.0)
    retries=state.get("eval_retries", 0)
    if faith >= FAITHFULNESS_THRESHOLD or retries >= MAX_EVAL_RETRIES:
        return "save"
    return "answer" 
```

```python
graph=StateGraph(CapstoneState)

graph.add_node("memory", memory_node)
graph.add_node("router", router_node)
graph.add_node("retrieve", retrieval_node)
graph.add_node("skip", skip_retrieval_node)
graph.add_node("tool", tool_node)
graph.add_node("answer", answer_node)
graph.add_node("eval", eval_node)
graph.add_node("save", save_node)

graph.set_entry_point("memory")

graph.add_edge("memory", "router")
graph.add_edge("retrieve", "answer")
graph.add_edge("skip", "answer")
graph.add_edge("tool", "answer")
graph.add_edge("answer", "eval")
graph.add_edge("save", END)

graph.add_conditional_edges("router", route_decision, {
    "retrieve": "retrieve",
    "skip": "skip",
    "tool": "tool"
})
graph.add_conditional_edges("eval", eval_decision, {
    "answer": "answer",
    "save": "save"
})

app=graph.compile(checkpointer=MemorySaver())
```

```python
def ask(question, thread_id="test"):
    print(f"\n{'='*55}")
    print(f"user: {question}")
    print(f"{'='*55}")
    config={"configurable": {"thread_id": thread_id}}
    result=app.invoke({"question": question, "eval_retries": 0}, config=config)
    print(f"bot: {result['answer']}")
    print(f"faithfulness: {result.get('faithfulness', 0):.2f} | route: {result.get('route')} | sources: {result.get('sources', [])}")
    return result["answer"]
```

```python
eval_dataset=[
    {
        "question": "How many days do I have to return an item?",
        "ground_truth": "30 days from the date of delivery"
    },
    {
        "question": "What is the shipping charge for orders below Rs. 499?",
        "ground_truth": "Rs. 49 standard shipping charge"
    },
    {
        "question": "Is Cash on Delivery available for all orders?",
        "ground_truth": "COD is available for orders up to Rs. 10000 for eligible pin codes with a Rs. 30 handling fee"
    },
    {
        "question": "How do I track my order?",
        "ground_truth": "Log in, go to My Orders, select the order and click Track Order. An SMS and email with tracking link is also sent after shipping."
    },
    {
        "question": "What happens if I miss the delivery?",
        "ground_truth": "Two more delivery attempts will be made on subsequent business days. After three failed attempts the order is returned and a refund is issued minus the shipping charge."
    }
]
ragas_results=[]
for item in eval_dataset:
    config={"configurable": {"thread_id": f"ragas_{hash(item['question'])}"}}
    result=app.invoke({"question": item["question"], "eval_retries": 0}, config=config)
    ragas_results.append({
        "question": item["question"],
        "answer": result["answer"],
        "contexts": [result.get("retrieved", "")],
        "ground_truth": item["ground_truth"],
        "faith_score": result.get("faithfulness", 0.0)
    })
    print(f"done: {item['question'][:50]}")

print("all 5 evaluation questions answered")
```

```python
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevancy, context_precision
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from ragas import RunConfig
from langchain_groq import ChatGroq
from langchain_huggingface import HuggingFaceEmbeddings
from datasets import Dataset

ragas_llm=LangchainLLMWrapper(ChatGroq(model="llama-3.3-70b-versatile", temperature=0))
ragas_embeddings=LangchainEmbeddingsWrapper(HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2"))

run_config=RunConfig(
    max_workers=1,
    max_retries=1,
    timeout=60
)

ds=Dataset.from_dict({
    "question": [r["question"] for r in ragas_results],
    "answer": [r["answer"] for r in ragas_results],
    "contexts": [r["contexts"] for r in ragas_results],
    "ground_truth": [r["ground_truth"] for r in ragas_results]
})

scores=evaluate(
    dataset=ds,
    metrics=[faithfulness, answer_relevancy, context_precision],
    llm=ragas_llm,
    embeddings=ragas_embeddings,
    run_config=run_config
)

df=scores.to_pandas()

print("="*40)
print(f"faithfulness:      {df['faithfulness'].mean():.3f}")
print(f"answer relevancy:  {df['answer_relevancy'].mean():.3f}")
print(f"context precision: {df['context_precision'].mean():.3f}")
print("="*40)
```

```python
import numpy as np
faith_scores=[r["faith_score"] for r in ragas_results]

def simple_relevancy(question, answer):
    question_words=set(question.lower().replace("?","").split())
    stopwords={"what","is","the","how","do","i","a","an","for","to","of","are","in","does"}
    keywords=question_words - stopwords
    answer_lower=answer.lower()
    if len(keywords)==0:
        return 1.0
    matches=sum(1 for word in keywords if word in answer_lower)
    return round(matches/len(keywords), 3)

def simple_precision(question, contexts, sources):
    if not contexts or not contexts[0]:
        return 0.0
    question_words=set(question.lower().split())
    context_words=set(contexts[0].lower().split())
    overlap=question_words & context_words
    return round(min(len(overlap)/max(len(question_words),1), 1.0), 3)

relevancy_scores=[]
precision_scores=[]

print("Manual Evaluation Results")
print("="*60)
for r in ragas_results:
    rel=simple_relevancy(r["question"], r["answer"])
    prec=simple_precision(r["question"], r["contexts"], r.get("sources",[]))
    relevancy_scores.append(rel)
    precision_scores.append(prec)
    print(f"Q: {r['question'][:50]}")
    print(f"   faithfulness:     {r['faith_score']:.2f}")
    print(f"   answer relevancy: {rel:.2f}")
    print(f"   context precision:{prec:.2f}")
    print()

print("="*60)
print("FINAL SCORES - copy these into your PDF")
print("="*60)
print(f"faithfulness:      {np.mean(faith_scores):.3f}")
print(f"answer relevancy:  {np.mean(relevancy_scores):.3f}")
print(f"context precision: {np.mean(precision_scores):.3f}")
print("="*60)
```
